````python
import numpy
import pandas
import matplotlib.pyplot as plt
import math
# Cost per click of individual keywords
b = 12 # try later 8 9 9.8
a = 50 # try later 50 40 44.5
def h(x):
 return b*x + a
xPlotValues=[0, 10]
x = [1.0, 2.1, 2.3, 2.5, 4.1, 4.5, 4.9, 5.9, 8.9]
#for i in range(0, 10):
#  x.append(i)
y = []
#for i in x:
#  y.append(h(i))
# Total amount of clicks per day
y = [48.2, 63.0, 89.0, 71.0, 89.0, 82.2, 70.0, 80.0, 150.0]


def Sum_Of_Squares(x, y, hFunc):
  result = 0
  # goes through all x, and y values in the x and y lists
  for numX, numY in zip(x,y):
      # calculates the difference between the actual and expected values
      dif = hFunc(numX) - numY
      # adds the difference squared to the result
      result = result + dif**2
  return result

print("r: " + str(Sum_Of_Squares(x, y, h)))
print()

plt.axis([0, 10, 0, 200])
plt.scatter(x, y)
regression_line = [h(item)for item in xPlotValues]
plt.plot([0, 10], regression_line)
````
